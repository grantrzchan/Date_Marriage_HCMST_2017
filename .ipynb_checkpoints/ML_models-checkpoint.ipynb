```python
#Married couples do not report the first year at which they met. Let's combine them as 'First_Together'
raw_data.insert(loc=1,column='First_Together',value=pd.Series(raw_data['w6_q21b_year'].copy()))
ind_list = list(raw_data[['w6_q21b_year','w6_q21d_year']][raw_data['w6_q21d_year'].notna()].index)
for i in ind_list:
        raw_data.at[i,'First_Together'] = raw_data.at[i,'w6_q21d_year']
    
#Couples for which the partner died do not report the year they break up. Let's combine them as 'End_Year'
raw_data.insert(loc=3,column='End_Year',value=pd.Series(raw_data['w6_q21e_year'].copy()))
ind_list = list(raw_data[['w6_q21e_year','w6_q21f_year']][raw_data['w6_q21f_year'].notna()].index)
for i in ind_list:
        raw_data.at[i,'End_Year'] = raw_data.at[i,'w6_q21f_year']
```

```python
raw_data.drop(columns=['w6_q21b_year', 'w6_q21d_year', 'w6_q21e_year', 'w6_q21f_year'],inplace=True)
raw_data.dropna(subset=['First_Together'],inplace=True)
```

```python
#create a column to figure out years together. Survey is taken in 2017, assume that N/A represents the couple is still dating
raw_data.insert(loc=1, column='Years_Together', value=pd.Series(raw_data['End_Year'].copy()))
ind_list = list(raw_data[raw_data['End_Year'].isna()].index)
for i in ind_list:
    raw_data.at[i,'Years_Together'] = 2017
raw_data['Years_Together'] = abs(raw_data['Years_Together'] - raw_data['First_Together'])
```

```python
cat_feats = list(raw_data.describe(include='category').columns)
num_feats = list(raw_data.describe(include='number').columns)
```

```python
#convert "Refused" in categorical features to NaN
treated_data = raw_data.copy().replace({'Refused': np.nan})
f_nans = treated_data.isna().sum()
print(f_nans)
```

```python
treated_data.shape[0]
```

- In order to keep as much usable data as possible, we should only allow a tolerance of missing data that is at most 5% of the total data collect. In this data set of 3330 responses, 5% represents approximately 166 responses.

- Therefore, let's remove columns that have more than 166 NaN values.

```python
for c in list(treated_data.columns):
    if treated_data[c].isna().sum() > 166:
        treated_data.drop(columns=[c], inplace=True)
print(treated_data.isna().sum())
```

```python
#use SimpleImputer from scikit-learn to deal with remaining NaN, using the "most frequent" strategy. "most frequent" is likely
#the most effective imputing method here because the largest percentage of NaN is relatively small (85/3330 equals 2.5%)
#and works well for both numeric and categorical data
from sklearn.impute import SimpleImputer as sim_imp
fill_nan = sim_imp(missing_values=np.nan, strategy='most_frequent')
imputed_data = pd.DataFrame(fill_nan.fit_transform(treated_data))
imputed_data.index = treated_data.index
imputed_data.columns = treated_data.columns
print(imputed_data.isna().sum())
```

```python
#convert categorical data to numeric form. Before deciding on whether to use one-hot encoding or ordinal encoding, let's
#examine the categorical features and their unique values

imputed_cat_feats = [c for c in imputed_data.columns if c in cat_feats]

for c in imputed_cat_feats:
    print(f'{c}: {len(tuple(imputed_data[c].unique()))}', sep='\n')
```

- Feature 'ppage' has an unusually high number of categories. It is likely a numeric feature misclassified as a categorical feature. Let's ignore 'ppage', and use one-hot encoding only with features that have unique values of 2 or less.

```python
# imputed_cat_feats.remove('ppage')

for c in imputed_cat_feats:
    d = {i[1]: i[0] for i in enumerate(tuple(imputed_data[c].unique()))}
    imputed_data[c].replace(d, inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split as tts
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error as mse
from math import sqrt
import eli5
from eli5.sklearn import PermutationImportance

oos_size = int(0.10*imputed_data.shape[0])
ins_size = imputed_data.shape[0] - oos_size
data = imputed_data[:ins_size]
oos_data = imputed_data[ins_size:]
```

```python
X = data.iloc[:, 3:].copy().to_numpy().astype('float64')
y = data.Years_Together.copy().to_numpy().astype('float64')
#create train test splits, set random_state = 0 to disable shuffling, test size is 25% of training examples
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.25, random_state=0)
```

```python
scores = []
rmses = []

rf_model = RandomForestRegressor(n_estimators= 1000, random_state=0).fit(X_train, y_train)
scores.append(rf_model.score(X_train, y_train))
y_hat = rf_model.predict(X_test)
rmses.append(sqrt(mse(y_test, y_hat, squared=True)))

perm = PermutationImportance(rf_model, random_state=1).fit(X_test, y_test)
eli5.show_weights(perm, feature_names = data.iloc[:, 3:].columns.tolist())
```
